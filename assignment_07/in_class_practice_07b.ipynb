```python
import pandas as pd

print("Pandas Version is ", pd.__version__)

import plotly 

print("Plotly Version is", plotly.__version__)

import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv("/content/wdi_data.csv")
df.head(10)
```

```python
df_gdp = df.query("indicator == 'NY.GDP.PCAP.PP.CD'")

df_gdp
```

```python
df_ind = df_gdp[df_gdp["Country Code"] == "IND"]

df_ind
```

```python
df_bra = df_gdp[df_gdp["Country Code"] == "BRA"]

df_bra
```

```python
fig = go.Figure()

trace_ind = go.Bar(      
    x=df_ind["Year"],    
    y=df_ind["value"],
    name="India"
)

fig.add_trace(trace_ind)


trace_brz = go.Bar(      
    x=df_bra["Year"],    
    y=df_bra["value"],
    name="Brazil"
)

fig.add_trace(trace_brz)

fig.update_layout(
    title="India vs Brazil GDP per Capita Over Time",
    xaxis_title="YEAR",
    yaxis_title="GDP per Capita"
)

fig.show()
```

```python
df_ind_bra = df_gdp[df_gdp["Country Code"].isin(["IND","BRA"])]

df_ind_bra
```

```python
fig = px.bar(
    data_frame=df_ind_bra,
    title="India GDP per Capita Over Time",
    labels={"value":"GDP per Capita", "Year":"YEAR"},
    orientation="v",
    x="Year",
    y="value",
    color="Country Code",
    barmode="group"
)

fig.update_xaxes(tickangle=45)


fig.show()
```
