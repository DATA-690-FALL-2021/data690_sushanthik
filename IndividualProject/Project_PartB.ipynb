```python
df = pd.read_csv("https://raw.githubusercontent.com/SushanthikPoreddy/data690_sushanthik/main/IndividualProject/wdi_data%20(3).csv")
df.head(10)
```

## GDP per capita by year in different countries:

```python
df_GDP = df.query('indicator == "NY.GDP.PCAP.PP.CD"')
df_GDP.head(10)
```

```python
df_GDP.drop(columns = ["Unnamed: 0"], inplace=True)
df_GDP.head(10)
```

```python
df_pivot1 = df_GDP.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'],columns = 'indicator', values ='value').reset_index()
df_pivot1.reset_index().head()
```

```python
fig = px.line(
    df_pivot1, 
    x="Year", 
    y="NY.GDP.PCAP.PP.CD",
    labels={"NY.GDP.PCAP.PP.CD":"GDP per capita, PPP (current international $)"},
    hover_name='Country Name',
    color='Country Name'
    )
fig.show()
```

## Mortality rate due to unsafe water, unsafe sanitation, and lack of hygiene in different countries:

```python
df_2016 = df.query("Year == 2016")
df_2016.sample(10)
```

```python
df_2016 = df_2016[df_2016["indicator"].isin(['SH.STA.WASH.P5'])]
df_2016.head()
```

```python
df_2016.drop(columns = ["Unnamed: 0","Year"], inplace=True)
df_2016.head(10)
```

```python
df_pivot = df_2016.pivot(index =['Country Code','Country Name','Region','Income Group','Lending Type'],columns = 'indicator', values ='value').reset_index()
df_pivot.reset_index().head()
```

```python
fig = px.bar(
    df_pivot,
    x="Country Name",
    y="SH.STA.WASH.P5",
    labels={"SH.STA.WASH.P5":"2016Mortality rate attributed to unsafe water, unsafe sanitation,<br> and lack of hygiene (per 100,000 population)"},
    color="Country Name"
)
fig.update_xaxes(categoryorder="total descending")
fig.show()
```

## How does mortality rate relate with total population and number of deaths that are caused by the communicable diseases?

```python
df_Deaths = df.query('indicator == "SH.DTH.COMM.ZS"')
df_Deaths.head(10)
```

```python
df_Deaths.drop(columns = ["Unnamed: 0"], inplace=True)
df_Deaths.head(10)
```

```python
df_pivot2 = df_Deaths.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'],columns = 'indicator', values ='value').reset_index()
df_pivot2.reset_index().head()
```

```python
fig = px.line(
    df_pivot2.dropna(),
    x="Year", 
    y="SH.DTH.COMM.ZS",
    color='Country Name',
    hover_name = "Country Name",
    labels={"SH.DTH.COMM.ZS":"Cause of death, by communicable diseases and maternal,<br> prenatal and nutrition conditions (% of total)"}
    )
#fig.update_xaxes(type='category')
fig.show()

### For this chart we see 2015.5, 2016.5 like that stuff between the years if we not use "fig.update_xaxes(type='category')".  
```
